```python
# first column dropped
main_df = main_df.drop(['By using this data you agree to the Terms of Use: https://www.ctdatacollaborative.org/terms-use'], axis=1)
```

```python
main_df.replace(['-99',-99], [np.nan, np.nan],inplace = True)
```

```python
# limiting df based on instance of records
threshold = 5000

# greater than threshold
limited_df = main_df.notnull().sum().where(main_df.notnull().sum() > threshold)
```

```python
limited_df.dropna()
```

```python
# less than threshold

limited_df = main_df.notnull().sum().where(main_df.notnull().sum() < threshold)
```

```python
limited_df.dropna()
```
